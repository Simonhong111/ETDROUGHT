```python
from epdatapro import *
from matplotlib import pyplot as plt
months =["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
```

```python
# 降雨异常
yy = "2018"
mm ="10"
chirpsdirectory = r"D:\Cornell\EthiopianDrought\Chirps2"
chirsparsdirectory = r"D:\Cornell\EthiopianDrought\AData\Chirps2Pars"

fig = plt.figure(figsize=(16,20))
plt.title("{} Rainfall Anomaly".format(yy), fontsize=20)
for idx,month in enumerate(months):
    ax = fig.add_subplot(6, 2, idx+1) # create 12-months subfigure
    ax.set_title(month)
    chirpsAnomMap = chirpsAnomalyMap(chirpsdirectory,chirsparsdirectory,yy,str(idx+1))
    mask = np.where(chirpsAnomMap > -9999)
    vmin = chirpsAnomMap[mask].min()
    vmax = chirpsAnomMap[mask].max()
    cax = ax.imshow(chirpsAnomMap,cmap=plt.get_cmap("RdBu"),vmin=vmin,vmax=vmax)
    cbar = plt.colorbar(cax,ax=ax) 
fig.tight_layout()#调整整体空白 
plt.show()
```

```python
# 降雨时间序列
chirpsdirectory = r"D:\Cornell\EthiopianDrought\Chirps2"

fig = plt.figure(figsize=(20,16))
plt.title("Rainfall Anomaly", fontsize=20)

for idx,month in enumerate(months):
    ax = fig.add_subplot(6, 2, idx+1) # create 12-months subfigure
    
    ave,year  = chirpsAnomalySeries(chirpsdirectory,str(idx+1))
    ax.plot(year,ave,label= months[idx])
    ax.grid(True)
    ax.legend()
    
    xmin, xmax, ymin, ymax = ax.axis()
    ax.vlines([2015,2016,2017], ymin, ymax,colors="r")
    ax.plot([xmin,xmax],[0,0],"k")
    ax.set_xticks(year)
    ax.set_xticklabels(year) 
    
    for xtick in ax.get_xticklabels():
        xtick.set_rotation(90)
fig.tight_layout()#调整整体空白 
plt.show()
```
